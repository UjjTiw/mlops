```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
train_df['Target']=encoder.fit_transform(train_df['Target'])
train_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(40, 20))
correlation_matrix = train_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(train_df, hue='Target')
plt.show()
```

```python
for column in train_df.columns:
    if column not in ['Target']:
        plt.figure(figsize=(12, 8))
        sns.boxplot(x='Target', y=column, data=train_df)
        plt.title(f'Box plot of {column} by status')
        plt.show()
```

```python
# After visualizing stuffs we conclude that we would need to not consider sfeatures whose correlation is less than 
from sklearn.ensemble import RandomForestClassifier

# Assume your features are in X and target in y
X = train_df.drop(columns=['id','Target'])
y = train_df['Target']

model = RandomForestClassifier()
model.fit(X, y)

feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False).plot(kind='bar')
plt.title('Feature Importance')
plt.show()
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score

# Assuming you have already split your data into training and testing sets
# X_train, X_test: Features, y_train, y_test: Target variable

# Define XGBoost classifier model
model = xgb.XGBClassifier()

# Train the model
model.fit(X, y)

# Predictions on the training set
train_predictions = model.predict(X)

# Evaluate accuracy on training set
train_accuracy = accuracy_score(y, train_predictions)
print("Training Accuracy:", train_accuracy)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define XGBoost classifier model
model = xgb.XGBClassifier()

# Define hyperparameters to search
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'objective': ['multi:softmax'],
    'num_class': [3]  # Number of classes
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=2, scoring='accuracy', n_jobs=-1)
grid_search.fit(X, y)

# Best parameters found during grid search
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Predictions on the test set using the best model
best_model = grid_search.best_estimator_
```

```python
test_df = pd.read_csv('/Users/ujjwaltiwari/Desktop/projects/mlops/experiment/playground-series-s4e6/test.csv')
# Assuming 'id' column is present in test_df
test_ids = test_df['id']
dicty = {0:'Dropout', 1:'Enrolled', 2: 'Graduate'}
# Drop 'id' column from test data if necessary
test_data = test_df.drop(columns=['id'])

# Make predictions using the trained model
test_predictions = best_model.predict(test_data)

# Create a DataFrame for predictions with 'id' and 'target' columns
predictions_df = pd.DataFrame({'id': test_ids, 'target': test_predictions})
predictions_df['target']= predictions_df['target'].map(dicty)

# Save predictions to CSV file
predictions_df.to_csv('sample1.csv', index=False)
```
